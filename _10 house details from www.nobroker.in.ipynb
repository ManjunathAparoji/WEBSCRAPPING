```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.nobroker.in/property/sale/bangalore/Malleshwaram?searchParam=W3sibGF0IjoxMy4wMDU1MTEzLCJsb24iOjc3LjU2OTIzNTgsInBsYWNlSWQiOiJDaElKdzg5NV8zODlyanNSdnFXa1JucXN1dDgiLCJwbGFjZU5hbWUiOiJNYWxsZXNod2FyYW0ifV0=&radius=2.0")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
house=soup.find_all('div', class_="nb__2JHKO")
house
```

```python
len(house)
```

```python
house.text
```

```python
house_text=[]
for i in house:
    for j in i.find_all('a', class_="nb__3CnI6"):
        house_text.append(j.text.replace("\n",""))
house_text
```

```python
len(house_text)
```

```python
location_text=[]
for i in house:
    for j in i.find_all('div', class_="nb__2CMjv"):
        location_text.append(j.text.replace("\n",""))
location_text
```

```python
len(location_text)
```

```python
emi=[]
for i in house:
    for j in i.find_all('div', class_="font-semi-bold heading-6"):
        emi.append(j.text.replace("\n",""))
emi
```

```python
emi_text=[]
for i in range(1,len(emi),3):
    emi_text.append(emi[i].strip())
emi_text
```

```python
len(emi_text)
```

```python
area_text=[]
for i in range(0,len(emi),3):
    area_text.append(emi[i].strip())
area_text
```

```python
len(area_text)
```

```python
price_text=[]
for i in range(2,len(emi),3):
    price_text.append(emi[i].strip())
price_text
```

```python
len(price_text)
```

```python
print(len(house_text), len(location_text), len(emi_text), len(area_text), len(price_text))
```

```python
import pandas as pd
House_Details=pd.DataFrame({})
House_Details['House Title']=house_text
House_Details['Location']=location_text
House_Details['Area']=area_text
House_Details['EMI']=emi_text
House_Details['Price']=price_text
```
