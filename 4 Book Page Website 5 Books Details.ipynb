```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
name=soup.find_all('div',class_="flex-article-content")
name
```

```python
len(name)
```

```python
name.text
```

```python
name_text=[]
for i in name:
    name_text.append(i.text)
name_text
```

```python
name_text=[]
for i in name:
    for j in i.find_all('h4', class_="italic"):
        name_text.append(j.text.replace("\n",""))
name_text
```

```python
len(name_text)
```

```python
author_text=[]
for i in name:
    for j in i.find_all('p', class_="sans bold"):
        author_text.append(j.text.replace("\n",""))
author_text
```

```python
len(author_text)
```

```python
genre_text=[]
for i in name:
    for j in i.find_all('p', class_="genre-links hidden-phone"):
        genre_text.append(j.text.replace("\n",""))
genre_text
```

```python
len(genre_text)
```

```python
reviews_text=[]
for i in name:
    for j in i.find_all('p', class_="excerpt"):
        reviews_text.append(j.text.replace("\n",""))
reviews_text
```
